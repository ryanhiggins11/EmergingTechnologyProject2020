```python
# Power Production Graph
import pandas as pd
import matplotlib.pyplot as plt

#Importing the dataset.
powerData = pd.read_csv("powerproduction.csv")
#Plotting the dataset.
plt.title('powerproduction.csv')
plt.xlabel('Speed')
plt.ylabel('Power')
plt.plot(powerData.values[:,0], powerData.values[:,1]);
```

```python
# Imports Used
import tensorflow.keras as kr
from tensorflow.python.keras.layers import Dense
from tensorflow.python.keras.models import Model
import numpy as np
np.set_printoptions(precision=3, suppress=True)
import pandas as pd
import matplotlib.pyplot as plt

#Reads in CSV file
powerData = pd.read_csv(
    "powerproduction.csv")

#Sets imported data set into Numpy arrays.
x = powerData.iloc[:,0].values
y = powerData.iloc[:,1].values

xArray = np.array(x)
yArray = np.array(y)

#Layers used to make the model
krInput = kr.Input(shape=(1,))
layer1 = Dense(10, activation='relu')(krInput)
layer2 = Dense(60, activation='relu')(layer1)
layer3 = Dense(60, activation='relu')(layer2)
layer4 = Dense(60, activation='relu')(layer3)
layer5 = Dense(60, activation='relu')(layer4)
layer6 = Dense(60, activation='relu')(layer5)
output = Dense(1)(layer6)

#This now compiles the created model together.
model = Model(inputs=krInput, outputs=[output])

#Uses adam optimizer and displays the loss when run.
model.compile(optimizer='adam', loss=['mean_squared_error'])

#This now fits the data arrays and speficies the epochs and batch size used.
model.fit(xArray, yArray,epochs=4000, batch_size=250)
model.predict([1.001])

#Saves model qhich is then used for making predictions.
model.save("Output_Model")
```

## Reasons i Used  this model:
I found that the fluctutation of the data was quite low to other models that i used. In comparison my last model i used the data was fluctuating from 200 to 300 which wasnt ideal. I decided to then turn to the realu layer as from researching it it is faster to compute the sigmoid function. It made a significant difference to training the nueral networks. As you can see from the data above the realu model here consistently acheived a loss from the 100s to 120s.
