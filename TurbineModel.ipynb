```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
import tensorflow.keras as kr
from tensorflow.keras import layers
from tensorflow.keras.layers.experimental import preprocessing


# Reading Data from powerproduction.csv
turbine_train = pd.read_csv(
    "powerproduction.csv",
    names=["speed", "power"])

turbine_features = turbine_train.copy()
turbine_labels = turbine_features.pop('power')

# Creates neural network with one neuron.
model = kr.models.Sequential()
model.add(kr.layers.Dense(15, input_shape=(1,), activation="sigmoid", kernel_initializer='ones', bias_initializer='zeros'))
model.add(kr.layers.Dense(125, input_shape=(1,), activation="sigmoid", kernel_initializer='ones', bias_initializer='zeros'))
model.add(kr.layers.Dense(1, activation="linear", kernel_initializer='ones', bias_initializer='zeros'))
model.compile('adam', loss='mean_squared_error')

# Train the neural network on our training data.
model.fit(turbine_features, turbine_labels, epochs=750)

#Saves data in model called output_model
model.save("Output_Model")
```
